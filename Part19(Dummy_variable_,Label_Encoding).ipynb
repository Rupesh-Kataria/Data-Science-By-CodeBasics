## Two method to create dummy variables one is pandas.get_dummies method and second was sklearn preprocessing one hot encoding library

```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/5_one_hot_encoding/homeprices.csv")
df
```

```python
dummies =pd.get_dummies(df.town)   #it will return me those dummy variable column . I should write categorical column inside that
dummies
```

```python
merged=pd.concat([df,dummies],axis='columns')
merged
```

```python
#we drop town column and one of the dummy variable column this is due to dummy variable trap
final=merged.drop(['town','west windsor'],axis='columns')
final
```

```python
from sklearn.linear_model import LinearRegression
model =LinearRegression()
```

```python
X=final.drop('price',axis='columns')
X
```

```python
y=final.price
y
```

```python
model.fit(X,y)
```

```python
#In real time situation you have lots of data so your training might takes a lot of time
#if you do not want to do training again and again . you can save model to a file and lode it back from file
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(X,y)  #it is used to check  how accurate our model is it is used to compare predicted y with actual y
```

```python
#Now to use one hot econder
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
```

```python
dfle=df
dfle.town=le.fit_transform(dfle.town)
dfle
```

```python
X=dfle[['town','area']].values   #I want X to be a two dimmensional array not a dataframe so I used values
X
```

```python
y=dfle.price
y
```

```python
#Now we need to create a dummy variable column 
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('one_hot_encoder', OneHotEncoder(),     #later whatever column I am supplying the 0th column have categorical_features
                         [0])], remainder='passthrough')
X = np.array(ct.fit_transform(X))
X
```

```python
#Now I have to remove one column due to that dummy variable trap
X=X[:,1:]
model.fit(X,y)
```

```python
model.predict([[1,0,2800]])
```
