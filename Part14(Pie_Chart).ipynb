```python
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet Bill","Car","Others Utilities"]
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
#to get rid of above text which is also shown along with pie diagram we use plt.show()
plt.pie(exp_vals,labels=exp_labels,radius=1.2)
plt.show()
```

```python
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.1f%%',shadow=True) #it is used to show upto 1 decimal value because in autopct we use 0.1
plt.show()
```

```python
#we can also use explode function which is used to take a part of pie out
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0]) #it is used to show upto 1 decimal value because in autopct we use 0.1
plt.show()
```

```python
#We can rotate the pie chart by using startangle
plt.pie(exp_vals,labels=exp_labels,radius=1.2,autopct='%0.1f%%',shadow=True,explode=[0,0.1,0.1,0,0],startangle=180)
plt.show()
```
