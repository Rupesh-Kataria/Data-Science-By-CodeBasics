```python
#pyplot is the main plot object that you get in matplotlib to plot chart
import matplotlib.pyplot as plt
%matplotlib inline  
```

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color="green",linewidth=5,linestyle='dashdot')
```

```python
#use of format string character
plt.plot(x,y,'g+') #Here g means green and + means + marker
```

```python
plt.plot(x,y,'g+--') #we also use -. in place of --
```

```python
#Now it does not matter in which order you specify the format string it in work every way
plt.plot(x,y,'--r*') #Here * is the marker ,used to mark the point
```

```python
#to use the diamond marker
plt.plot(x,y,'rD')
```

```python
#Now if you do not like the format string then you can also pass the individual argument
plt.plot(x,y,color='red',marker='D',linestyle='')
```

```python
#for the color you can also specify the hexadecimal string and you can also specify the marker size
plt.plot(x,y,color='#008040',marker='D',linestyle='',markersize=20)
```

```python
#alpha is use for color transperency its value is between 0 to 1
plt.plot(x,y,color="green",alpha=0.2)
```
