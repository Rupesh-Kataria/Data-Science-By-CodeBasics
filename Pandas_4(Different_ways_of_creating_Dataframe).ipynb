## Different Ways of Creating Dataframe 
To know more about them go to pandas io which consist of API to read and write


```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/3_different_ways_of_creating_dataframe/weather_data.csv")
df
```

```python
df=pd.read_excel("weather_data.xlsx","Sheet1")
df
```

```python
weather_data = {                                   #in the above dictionary method
    'day': ['1/1/2017','1/2/2017','1/3/2017'],
    'temperature': [32,35,28],
    'windspeed': [6,7,2],
    'event': ['Rain', 'Sunny', 'Snow']
}
df=pd.DataFrame(weather_data)
df
```

```python
weather_data = [                         #this is a tuple of rows 
    ('1/1/2017',32,6,'Rain'),
    ('1/2/2017',35,7,'Sunny'),
    ('1/3/2017',28,2,'Snow')
]
#Note while creating dataframe using tuple you need to give the name of column
df=pd.DataFrame(weather_data,columns=["day","temperature","windspeed","event"])
df
```

```python
weather_data = [          #dictionary inside list
    {'day': '1/1/2017', 'temperature': 32, 'windspeed': 6, 'event': 'Rain'},
    {'day': '1/2/2017', 'temperature': 35, 'windspeed': 7, 'event': 'Sunny'},
    {'day': '1/3/2017', 'temperature': 28, 'windspeed': 2, 'event': 'Snow'},
    
]
df=pd.DataFrame(weather_data)
df
```
