```python
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/6_handling_missing_data_replace/weather_data.csv")
df
```

```python
new_df=df.replace([-99999,-88888],np.NaN) #replace all the special value that is -99999,-88888 with NaN
new_df
```

```python
#if I also want to replace event 0 value with NaN but i can not do with the above method because it replace 0 in all column with NaN but I do not want that so we have to use the dictionary method

new_df=df.replace({
    'temperature':-99999,
    'windspeed':-99999,
    'event':'0'       #Note we should not write 0 because it is a string so we write it in close bracket
    },np.NaN) 
new_df
```

```python
#we can also use other mapping method to replace -99999 with NaN and 'Sunny' with 'Rainy' by using mapping this time
new_df=df.replace({
    -99999:np.NaN,
    'Sunny':'Rainy'
})
new_df
```

```python
#Note what if the value in temperature column,windspeed column are in their unit like C(celsisus),F(faranehit) if we want to remove them then we use regex method
#Regex method is used to replace alphatetical value like (unit)
#Although in the above they are not provided in their unit
#In the below '' means blank value it means it replace the dimension value with blank value
new_df=df.replace({
    'temperature':'[A-Za-z]',
    'windspeed':'[A-Za-z]'
},'',regex=True)
new_df
```

```python
#We can also use this below method but the problem is that it replace the entire column of rain  with  '' blank because it is alphabetical that's why we need dictionary method
new_df=df.replace('[A-Za-z]','',regex=True)
new_df
```
