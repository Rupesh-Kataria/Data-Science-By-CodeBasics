```python
import pandas as pd
india_weather =pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30],
    "humidity":[80,60,78]
})
india_weather
```

```python
us_weather=pd.DataFrame({
    "city":["new york","chicago","orLando"],
    "temperature":[21,14,35],
    "humidity":[68,65,75]
})
us_weather
```

```python
#two concat two dataframes 
df=pd.concat([india_weather,us_weather])
df
```

```python
#if we want continuous index then we can pass an argument ignore index
df=pd.concat([india_weather,us_weather],ignore_index=True)
df
```

```python
#we can also create another index apart from numerical index like country index
df=pd.concat([india_weather,us_weather],keys=["india","us"])
df
```

```python
df.loc["us"]
```

```python
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30]
})
temperature_df
```

```python
windspeed_df=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "windspeed":[7,12,9]
})
windspeed_df
```

```python
df=pd.concat([temperature_df,windspeed_df],axis=1)
df
```

```python
#What if the cities are not in same order in temperatur_df and windspeed_df then we use to provide index to them to concate propely
temperature_df=pd.DataFrame({
    "city":["mumbai","delhi","banglore"],
    "temperature":[32,45,30]
},index=[0,1,2])
temperature_df
```

```python
#Now we assign delhi as index 1 in windspeed_df because it has index one in temperature_df to conconate them properly
windspeed_df=pd.DataFrame({
    "city":["delhi","mumbai"],
    "windspeed":[7,12]
},index=[1,0])
windspeed_df
```

```python
df=pd.concat([temperature_df,windspeed_df],axis=1)
df
```
