```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/pandas/5_handling_missing_data_fillna_dropna_interpolate/weather_data.csv")
df
```

```python
df.to_csv("weather_data.csv") 
df1=pd.read_csv("weather_data.csv")
type(df1.day[0])
```

```python
#since day is in string form we want to convert it into date(time form so we have to parse it)
df1=pd.read_csv("weather_data.csv",parse_dates=["day"])
type(df1.day[0])
```

```python
#Now if i want to set day as index
df1.set_index('day',inplace=True)
df1
```

```python
new_df=df.fillna(0)    #filling NaN value with 0
new_df
```

```python
#if we do not want to fill all dataframe NaN value with 0 then we can use dictionary method 
#Now we can see below the temperature and windspeed is replaced with 0  and wind speed with no event
new_df=df.fillna({
    'temperature':0,
    'windspeed':0,
    'event':'no event'
})
new_df
```

```python
new_df=df.fillna(method="ffill") #ffill means if we have NaN value carry forward the previous value
new_df
```

```python
#ffill means forward fill value we can also we use bfill backward fill value that is we can fill next day value instead of previous day value
new_df=df.fillna(method="bfill",axis="columns") #it fill horizontally
new_df
```

```python
new_df=df.fillna(method="ffill",limit=1)   #it mean we can fill above element to upto 1 value to NaN
new_df
```

```python
new_df=df.interpolate() #filling intermediate value in between  #linear interpolation
new_df
```

```python
#In above intepolate method it consider only mean value it do not consider the time value to get relative average
#so we need to get time weighted interpolation but for that we have to convert the time as index
new_df=df1.interpolate(method="time")
new_df
```

```python
#drop all the row with NaN values
new_df=df.dropna()
new_df
```

```python
# If I want to drop only that  row which have all NaN values considering date as index and leaving index
new_df=df1.dropna(how="all")
new_df
```

```python
#if I have atleast one not  NaN value then keep that row that so for that we shoul set threshold=1
new_df=df1.dropna(thresh=1) #so i should keep that row which have atleast one valid value we do not count day in that because it is index
new_df
```

```python
#If we want to insert all mid dates in day then we use this below method 
dt=pd.date_range("01-01-2017","01-11-2017")
idx=pd.DatetimeIndex(dt)
df2=df1.reindex(idx)
df2
```
