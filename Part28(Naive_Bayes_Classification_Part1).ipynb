```python
import pandas as pd
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/14_naive_bayes/titanic.csv")
df.head()
```

```python
#I am gonn a drop some variable which do not have any impact like name 
df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns',inplace=True)
df.head()
```

```python
target=df.Survived
inputs=df.drop('Survived',axis='columns')
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
inputs=pd.concat([inputs,dummies],axis='columns')
inputs.head(3)
```

```python
inputs.drop('Sex',axis='columns',inplace=True)   #it is simple with no text and only number
inputs.head(3)
```

```python
#I also want to find if there is na values in any columns
inputs.columns[inputs.isna().any()]
```

```python
inputs.Age[:10]
```

```python
inputs.Age = inputs.Age.fillna(inputs.Age.mean())#fill means value in place of na at age
inputs.head(6)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(inputs,target,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
len(inputs)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB() #WE USE THIS WHEN OUR DATA DISTRIBUTION IS NORMAL OR GAUSIAN
```

```python
#our trainning become very fast because our sample is less .When sample is high it takes some time
model.fit(X_test,y_test)
```

```python
model.score(X_test,y_test) #our score will vary on every run of the program
```
