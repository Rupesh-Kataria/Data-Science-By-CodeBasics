```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend() #it is used to show which line belong to what  #if i want to put my legends in upper right corner i used  plt.legend(loc="upper right")
```

```python
plt.xlabel("Days")
plt.ylabel("Temperature")
plt.title("Weather")
plt.plot(days,max_t,label="Max")
plt.plot(days,min_t,label="Min")
plt.plot(days,avg_t,label="Avg")
plt.legend(loc="best",shadow=True) #it move legend to best  position that is where it do not cross the line and shadow used to create shadow
plt.grid()  #sometime people used grid to visualise property best 
```
