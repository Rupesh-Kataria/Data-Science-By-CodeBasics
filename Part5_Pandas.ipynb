## Now seeing the xl file

```python
#Now reading the xl file and making some changes with the help of xl
df=pd.read_excel("stock_data.xlsx","Sheet1")

def convert_people_cell(cell):
  if cell=="n.a.":
      return 'sam walton'
  return cell

def convert_eps_cell(cell):
  if cell=="not available":
       return None
  return cell

df= pd.read_excel("stock_data.xlsx","Sheet1",converters={   #it go the each cell of people and called the function and changed the value if it is 'n.a.'
            'people':convert_people_cell,
            'eps':convert_eps_cell
})
df
```

```python
#converting the above dataframe into xl file
df.to_excel("new.xlsx",sheet_name="stocks")
```
